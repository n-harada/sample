```python
import os
import pandas as pd
import pickle
import re
import jaconv

root_dir= '/home/yukim/Desktop/prescription/'
out_dir = '/home/yukim/Desktop/prescription/DB_module_pickles'
root_data_dir = os.path.join(root_dir,'mzd_for_upload/update_hospital_DB_and_symspell/hospital_data_raw_of_20201115')
```

```python
dict_medical_place_type = {'医科':'1','歯科':'3','訪問':'6'} #医科併設,歯科併設という変化球パターンに未対応なことに留意

dict_place_code = {'茨城':'08','栃木':'09','群馬':'10','埼玉':'11',
                   '千葉':'12','東京':'13','神奈川':'14',
                   '新潟':'15','山梨':'19','長野':'20'}
```

```python
os.chdir(out_dir)
with open('dict_hospitalInfo.pickle','rb') as f:
    dict_OLD = pickle.load(f)
dict_OLD
```

```python
os.chdir(root_data_dir)
list_file_dir = os.listdir()

dict_out = {}
for dir_ in list_file_dir:
    dir_ = os.path.join(root_data_dir,dir_)
    print(dir_)
    os.chdir(dir_)
    list_dir = os.listdir()
    for i in range(len(list_dir)):
        filename = list_dir[i]
        print(filename)

        #今回は取得しない情報はここで弾く
        if '歯科' in filename or '薬局' in filename:
            print(filename,'passed')
            continue
        
        #ファイル名より, 医療機関種別・医療機関都道府県コードを取得する
        medical_spot_type, medical_spot_place_code = '0','0'
        cnt = 0
        for key in dict_medical_place_type.keys():
            if key in filename:
                medical_spot_type = dict_medical_place_type[key]
                cnt+=1
                if cnt>1:
                    raise ValueError('there was overlap in place code')

        cnt = 0
        for key in dict_place_code.keys():
            if key in filename:
                medical_spot_place_code = dict_place_code[key]
                cnt+=1
                if cnt>1:
                    raise ValueError('there was overlap in place name')

        #excelを読み込み, 医療機関名称・医療機関コードを取得する
        ##read
        df = pd.read_excel(filename)
        df.columns = list(range(len(df.columns)))

        ##column 2に入っている項番がある行を抽出する
        df[2] = df[2].fillna(-1)
        l_with_kouban = []
        for i in df.index:
            if df.at[i,2] >=0:
                l_with_kouban.append(i)
        df = df.loc[l_with_kouban].set_index(2,drop=True)

        for i in df.index:
            nums = df.at[i,6].splitlines()[0].split('(')[0]
            medical_spot_code = ''.join(re.findall('[0-9]+',nums))
            name = jaconv.normalize(df.at[i,10])

            if len(medical_spot_code)!=7:
                raise ValueError(name,nums,medical_spot_code,'medical spot code wrong')

            l_vals = [medical_spot_type,medical_spot_code,medical_spot_place_code]
                
            dict_out[name] = [jaconv.normalize(i) for i in l_vals]
```

```python
#~~法人 or ~~会が入っている病院は, その接頭語を除いた文字列も病院名として用いる
l_header = ['法人','会','組合','付属','施設','保険','付属','直営','保健所']

l_hospital_headers = []
for name in list(dict_out.keys())[:]:
    
    l_split = [i for i in name.split(' ') if i!=''] #''がl_splitに含まれる事象を確認したためPATCH
    header_found=True
    while len(l_split)>1 and header_found==True:
            
        #判定対象の文字列を定義する
        frontier_val = l_split[0]
            
        #医療法人~~ や ~~会, は空白にてsplitされているという前提のもと, その接頭語を落としていく
        header_found = False
        for h in l_header:
            if h in frontier_val:
                header_found=True
        if header_found:
            l_hospital_headers.append(l_split[0])
            l_split = l_split[1:]
            
    abbr = ''.join(l_split)
    if name!=abbr:
        dict_out[abbr] = dict_out[name][:]
```

```python
os.chdir(root_dir)
with open('dict_hospitalInfo_NEW.pickle','wb') as f:
    pickle.dump(dict_out,f)
with open('l_hospital_headers_NEW.pickle', 'wb') as f:
    pickle.dump(l_hospital_headers, f)
```
