```python
import os
import pandas as pd
import pickle
import jaconv
import itertools
from tqdm.notebook import tqdm_notebook as tqdm
```

### 大本の結果集計

```python
os.chdir('/home/yukim/Desktop/symspell_local/results')
with open('df_result_full.pickle', 'rb') as f:
    df_result_base = pickle.load(f)
```

```python
#正解データの作成
l1 = ['イトラコナゾール錠50mg「科研」','ルリコンクリーム1%','ベギンクリーム20%','ベピオゲル2.5%']
l2 = ['正解不明','正解不明','正解不明','ユーパスタコーワ軟膏','ビラノア錠20mg']
l3 = ['シナール配合錠','フラビタン錠10mg','ピドキサール錠20mg','ジスロマック錠250mg']
l4 = ['ヒルドイドソフト軟膏0.3%','ビーソフテンローション0.3%','マイザー軟膏0.05%','プロペト','プロトピック軟膏0.1%','プロペト']
l5 = ['スタデルム軟膏5%','ファムビル錠250mg','メチコバール錠500μg','ロキソニン錠60mg','カロナール錠300','ムコスタ錠100mg']
l6 = ['イトラコナゾール錠50mg「科研」','ルリコンクリーム1%','ベギンクリーム20%','ベピオゲル2.5%']
l_phc = ['カルボシステイン錠250mg','アンブロキソール塩酸塩錠15mg','フスコデ配合錠','セフカペンピボキシル塩酸塩錠100mg','ニフェジピン徐放錠20mg(24時間持続)','ケトプロフェンテープ40mg(10×14cm非温感)']

l_corrects = [l1,l2,l3,l4,l5,l6,l_phc]
```

```python
#広義の正解数・狭義の正解数・ノイズ数、を導出
#memo: 理想は狭義の正解数が高く, ノイズ数が低いパラメータ
df_result_all = df_result_base.copy()
df_result_all = df_result_all.assign(l_correct=None,total_picked=0,broad_correct_num=0,narrow_correct_num=0,
                            broad_precision=0.0,broad_recall=0.0,broad_f=0.0,
                             narrow_precision=0.0,narrow_recall=0.0,narrow_f=0.0)

for i in tqdm(range(len(df_result_all))):
    
    #正解の定義・計算結果の取得
    try:
        sample_num = int(df_result_all.at[i,'input'].split('_')[0][-1])
        l_correct = l_corrects[sample_num-1]
    except ValueError:
        l_correct = l_corrects[-1]
        
    l_calced_vals = df_result_all.at[i,'calced_vals']
    
    #全部の取得数
    n_all = len(l_calced_vals)
    
    #1つも医薬品を取得していない場合は、continueする
    if n_all==0:
        continue
    
    #広義・狭義の正解数導出
    broad_correct,narrow_correct = 0,0
    for correct in l_correct:
        searching_broad,searching_narrow = True,True
        for small_list in l_calced_vals:
            if (correct in small_list[-8:]) & searching_broad:
                broad_correct+=1
                searching_broad=False
            if (correct==small_list[-1]) & searching_narrow:
                narrow_correct+=1
                searching_narrow=False
                
    if n_all-broad_correct<0:
        break
        
    df_result_all.at[i,'l_correct'] = l_correct
    df_result_all.at[i,'total_picked'] = n_all
    
    df_result_all.at[i,'broad_correct_num'] = broad_correct    
    precision = broad_correct/len(l_calced_vals)
    recall = broad_correct/len(l_correct)
    df_result_all.at[i,'broad_precision'] = precision
    df_result_all.at[i,'broad_recall'] = recall
    df_result_all.at[i,'broad_f'] = (2*recall*precision)/(recall+precision)
    
    df_result_all.at[i,'narrow_correct_num'] = narrow_correct    
    precision = narrow_correct/len(l_calced_vals)
    recall = narrow_correct/len(l_correct)
    df_result_all.at[i,'narrow_precision'] = precision
    df_result_all.at[i,'narrow_recall'] = recall
    df_result_all.at[i,'narrow_f'] = (2*recall*precision)/(recall+precision)
```

```python
df_result_all = df_result_all.fillna(0.0)
```

### 結果の分割・分析

```python
df = df_result_all.copy()
df_result_clear = df[(df.input=='sample1_med.txt')|(df.input=='sample3_med.txt')|(df.input=='sample4_med.txt')|(df.input=='sample6_med.txt')|(df.input=='PHC_sample_med.txt')]
df_result_ambiguous = df[(df.input=='sample2_med.txt')|(df.input=='sample5_med.txt')]
```

```python
##全体に対してF値をaggする
df_result = df_result_all.copy()

#agg result
group = df_result.groupby(['max_edit_distance_dict','prefix','val_len_min','val_len_max','max_edit_distance_lookup'])
df_agg = group.mean()[['broad_f','narrow_f','time']]

#普通にf値でソート
df_agg.sort_values('val_len_min')
df_agg.sort_values(['broad_f','narrow_f','max_edit_distance_dict','max_edit_distance_lookup','prefix','val_len_max',],ascending=False)[:20]
#df_agg.sort_values(['narrow_f','broad_f','max_edit_distance_dict','prefix','val_len_max',],ascending=False)
```

```python
##高精度な入力のみに対してF値をaggする

df_result = df_result_clear.copy()

#agg result
group = df_result.groupby(['max_edit_distance_dict','prefix','val_len_min','val_len_max','max_edit_distance_lookup'])
df_agg = group.mean()[['broad_f','narrow_f','time']]

#普通にf値でソート
df_agg.sort_values('val_len_min')
df_agg.sort_values(['broad_f','narrow_f','max_edit_distance_dict','max_edit_distance_lookup','prefix','val_len_max',],ascending=False)[:20]
#df_agg.sort_values(['narrow_f','broad_f','max_edit_distance_dict','prefix','val_len_max',],ascending=False)
```

```python
##ぶれている入力のみに対してF値をaggする

df_result = df_result_ambiguous.copy()

#agg result
group = df_result.groupby(['max_edit_distance_dict','prefix','val_len_min','val_len_max','max_edit_distance_lookup'])
df_agg = group.mean()[['broad_f','narrow_f','time']]

#普通にf値でソート
df_agg.sort_values('val_len_min')
df_agg.sort_values(['broad_f','narrow_f','max_edit_distance_dict','max_edit_distance_lookup','prefix','val_len_max',],ascending=False)[:20]
#df_agg.sort_values(['narrow_f','broad_f','max_edit_distance_dict','prefix','val_len_max',],ascending=False)
```
