```python
# 基本は最頻値をとる。最頻値がない時は平均をとる。numsが空の時は0にする。
def mode_average(nums):
    try:
        return mode(nums)
    except:
        if len(nums)!=0:
            return round(sum(nums) / len(nums))
        else:
            try:
                return nums[0]
            except:
                return 0  #これは危ないかも
```
