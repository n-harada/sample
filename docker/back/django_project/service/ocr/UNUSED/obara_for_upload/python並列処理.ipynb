```python
def path_to_array(img_path):
    '''
    画像pathを画像arrayに変換する
    input:画像path
    output:array
    '''
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```
